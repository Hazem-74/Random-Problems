# Harmonic Oscillator

```python
import sympy as sp
from IPython.display import display, Math

# Define symbols
x, m, omega, hbar, n = sp.symbols("x m omega hbar n", real=True, positive=True)

# Schrödinger equation components
psi = sp.Function("psi")(x)  # Wavefunction
V = (1/2) * m * omega**2 * x**2  # Potential energy
E = sp.Symbol("E", real=True)  # Energy eigenvalue
```

```python
# Time-independent Schrödinger equation
schrodinger_eq = (-hbar**2 / (2 * m) * sp.diff(psi, x, 2) + V * psi - E * psi).simplify()

# Display the Schrödinger equation
display(Math(r"\textbf{Schrödinger Equation:}"))
display(Math(sp.latex(schrodinger_eq) + r" = 0"))
```

```python
from sympy import symbols, Function, sqrt, factorial, pi, exp, Rational
from sympy.abc import n, x
from sympy.printing import latex

# Define constants and variables
m, omega, hbar = symbols('m omega hbar', positive=True, real=True)
xi = symbols('xi')
Hn = Function('H')(xi)  # Hermite polynomial

# Define the wavefunction
psi = ((m * omega / (pi * hbar))**Rational(1, 4)) * (1 / sqrt(2**n * factorial(n))) * Hn * exp(-xi**2 / 2)

# Print LaTeX representation
from IPython.display import display, Math
display(Math(latex(psi)))
```

```python
# Energy eigenvalues and eigenfunctions
k = m * omega / hbar  # Simplified constant
norm_factor = sp.sqrt(sp.sqrt(k) / (sp.factorial(n) * 2**n * sp.sqrt(sp.pi)))
H_n = sp.hermite(n, sp.sqrt(k) * x)  # Hermite polynomial
psi_n = norm_factor * sp.exp(-k * x**2 / 2) * H_n  # Normalized eigenfunction
E_n = hbar * omega * (n + 1/2)  # Energy eigenvalue

# Display normalized eigenfunction
display(Math(r"\textbf{Normalized Eigenfunction:}"))
display(Math(r"\psi_n(x) = " + sp.latex(psi_n)))
```

```python
psi_n = sp.simplify(psi_n)
# Display normalized eigenfunction
display(Math(r"\textbf{Normalized Eigenfunction:}"))
display(Math(r"\psi_n(x) = " + sp.latex(psi_n)))
```

```python
# Display energy eigenvalue
display(Math(r"\textbf{Energy Eigenvalue:}"))
display(Math(r"E_n = " + sp.latex(E_n)))
```

```python
# Importing necessary libraries
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.special import hermite

# Define real-world constants
hbar = 1.0545718e-34  # Reduced Planck's constant in J·s
m = 9.10938356e-31    # Mass of an electron in kg
omega = 1  # Angular frequency for a frequency of  Hz (in rad/s)

# Define a numerical function to evaluate the eigenfunction at specific x values
def psi_n_numeric(n_val, x_vals):
    """Numerically evaluate the eigenfunction at specific x values."""
    norm_factor_val = np.sqrt(np.sqrt(m * omega / hbar) / (math.factorial(n_val) * 2**n_val * np.sqrt(np.pi)))
    H_n_vals = hermite(n_val, monic=True)(np.sqrt(m * omega / hbar) * x_vals)
    psi_vals = norm_factor_val * np.exp(- (m * omega / (2 * hbar)) * x_vals**2) * H_n_vals
    return psi_vals
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite, factorial

# Define constants
m = 1.0  # mass (arbitrary units)
omega = 1.0  # angular frequency (arbitrary units)
hbar = 1.0  # Planck's constant (arbitrary units)

# Define the wavefunction
def psi_n(x, n):
    xi = np.sqrt(m * omega / hbar) * x
    normalization = (m * omega / (np.pi * hbar))**0.25 / np.sqrt(2**n * factorial(n))
    Hn = hermite(n)  # Hermite polynomial
    return normalization * Hn(xi) * np.exp(-xi**2 / 2)

# Set up the plot
x = np.linspace(0, 5, 1000)  # x range
n_values = list(range(0, 10))

plt.figure(figsize=(10, 6))
for n in n_values:
    plt.plot(x, psi_n(x, n), label=fr"$n={n}$")  # Use raw string for LaTeX

# Customize the plot
plt.title(r"Wavefunctions $\psi_n(x)$ for Harmonic Oscillator", fontsize=14)  # Raw string
plt.xlabel(r"$x$", fontsize=12)  # Raw string
plt.ylabel(r"$\psi_n(x)$", fontsize=12)  # Raw string
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite, factorial

# Define constants
hbar = 1.0545718e-34  # Reduced Planck's constant in J·s
m = 9.10938356e-31    
omega = 3.0  # angular frequency (arbitrary units)

# Define the wavefunction
def psi_n(x, n):
    xi = np.sqrt(m * omega / hbar) * x
    normalization = (m * omega / (np.pi * hbar))**0.25 / np.sqrt(2**n * factorial(n))
    Hn = hermite(n)  # Hermite polynomial
    return normalization * Hn(xi) * np.exp(-xi**2 / 2)

# Define the probability density function
def P_n(x, n):
    return np.abs(psi_n(x, n))**2

# Set up the plot
x = np.linspace(-0.1, 0.1, 1000)  # x range
n_values = list(range(100, 101))

plt.figure(figsize=(10, 6))
for n in n_values:
    plt.plot(x, P_n(x, n), label=f"$n={n}$")

# Customize the plot
plt.title("Normalized Probability Distribution $P_n(x)$", fontsize=14)
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$P_n(x)$", fontsize=12)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid()
plt.savefig('Harmonic.png')
plt.show()
```
