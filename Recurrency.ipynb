```python
from sympy import symbols, Function, diff, simplify,exp, summation, oo
from IPython.display import display, Latex
```

```python
E, mu, K_b, T, g_n = symbols('E mu K_b T g_n')
l, n = symbols('l n', integer=True)
Z = Function('Z')(T)
```

```python
Z_expr  = 1/(1 - exp(-(E-mu)/(K_b*T)))
```

```python
Z_expr
```

```python
g_n = summation(Z_expr**l / l**n, (l, 0, oo))
```

```python
g_n
```

```python
dTg_n = diff(g_n,T)
dTg_n
```

```python
for i in range(10):
    dTZ_expr = simplify(diff(Z_expr,T))
    Z_expr = dTZ_expr
    display((dTZ_expr))
```

```python
dTg_n.subs(Z_expr,dTZ_expr)
dTg_n
```
