```python
from sympy import sin, cos, DiracDelta ,latex, symbols, integrate,oo
from IPython.display import Math
x, alpha, beta = symbols('x alpha beta')
```

```python
f = integrate(DiracDelta(x - alpha)*DiracDelta(x - beta), (x,-oo,oo)) 
Math(r'\int_{-\infty}^\infty \delta(x - \alpha) \delta(x - \beta) \, dx = ' + latex(f))
```

```python
g = integrate(DiracDelta(x - alpha)**2, (x,-oo,oo)) 
Math(r'\int_{-\infty}^\infty \delta(x - \alpha)^2 \, dx = ' + latex(g))
```
